```python
def get_uniques(df,columns):
    return {column:list(df[column].unique()) for column in columns}
```

```python
categorical_columns=['region','Date','type']
get_uniques(avocado1_dataset,categorical_columns)
```

```python
ordinal_features=['Date']
nominal_features=['region']
target_column='type'
```

```python
Date_ordering=sorted(avocado1_dataset['Date'].unique())
```

```python
def ordinal_encode(df,column,ordering):
    df=df.copy()
    df[column]=df[column].apply(lambda x:ordering.index(x))
    return df
    
    
    
def onehot_encoder(df,column):
    df=df.copy()
    dummies=pd.get_dummies(df[column])
    df=pd.concat([df,dummies],axis=1)
    df.drop(column,axis=1,inplace=True)
    return df
```

```python
avocado1_dataset=ordinal_encode(avocado1_dataset,'Date',Date_ordering)
```

```python
avocado1_dataset=onehot_encoder(avocado1_dataset,'region')
```

```python
avocado1_dataset
```

```python
label_encoder=LabelEncoder()
avocado1_dataset[target_column]=label_encoder.fit_transform(avocado1_dataset[target_column])
```

```python
y=avocado1_dataset[target_column]
x=avocado1_dataset.drop(target_column,axis=1)
```

```python
Scaler=StandardScaler()
x=Scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
```

```python
inputs=tf.Keras.Input(66)
x=tf.Keras.layers.Dense(65,activation='relu')(inputs)
x=tf.Keras.layers.Dense(65,activation='relu')(x)
outputs=tf.Keras.layers.Dense(1,activation='sigmoid')(x)
model=tf.Keras.Model(inputs=inputs,outputs=outputs)

model.compile(
optimizer='adam',
loss='mse',
metrics=['accuracy']
)
history=model.fit(
x_train,
y_train,
vaildation_split=0.2,
batch_size=64,
epochs=100,
callbacks=[tf.Keras.callbacks.ReduceLROnPlateau()],
verbose=0
)
```
